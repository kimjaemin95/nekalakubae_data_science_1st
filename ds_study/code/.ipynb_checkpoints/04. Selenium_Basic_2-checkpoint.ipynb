```python
import pandas as pd
import time
from selenium import webdriver
```

```python
url = "https://finance.naver.com/marketindex/"
driver = webdriver.Chrome("../driver/chromedriver.exe")
driver.get(url)
```

```python
# 환전 고시 환율
contents = driver.find_elements_by_css_selector("#exchangeList > li > a")
contents[0].find_element_by_css_selector(".value").text
```

```python
# 주요 뉴스 제목
news = driver.find_elements_by_css_selector("#content > div.section_news > div > ul > li:nth-child(1) > p > a")
news[0].text
```

```python
# 페이지 화면 화단 매매기준율 테이블 정보(iframe)
driver.find_elements_by_css_selector("body > div > table > tbody > tr")
```

## iframe 접근

```python
# selenium을 이용한 접근
```

```python
# iframe 태그 지정

frame_ex1 = driver.find_element_by_css_selector("#frame_ex1")
frame_ex1
```

### iframe 태그로 이동

```python
# iframe 태그로 이동
driver.switch_to.frame(frame_ex1)
```

```python
# 밑 페이지에 있는 매매기준율
contents = driver.find_elements_by_css_selector("body > div > table > tbody > tr")
contents[0].find_element_by_css_selector(".sale").text
```

```python
datas = []

for cont in contents:
    title = cont.find_element_by_css_selector(".tit").text
    link = cont.find_element_by_css_selector(".tit > a").get_attribute("href")
    sale = cont.find_element_by_css_selector(".sale").text
    print(title, sale, url)
    datas.append({
        "title":title,
        "sale":sale,
        "link":link
    })
```

```python
df = pd.DataFrame(datas)
df.head()
```

```python
df.to_excel("../data/sel_naver_finance.xlsx", encoding="utf-8")
```

```python
driver.close()
```

### requests를 이용한 호출

```python
import requests
from bs4 import BeautifulSoup

url = "https://finance.naver.com/marketindex/exchangeList.naver"
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
soup.find("td", class_="tit").text
```

```python
soup.find("td", class_="sale").text
```

```python
contents = soup.select("tbody > tr")
for content in contents:
    print(content.select_one(".tit").text.strip(),content.select_one(".sale").text, sep="  \t|\t  ")
```
