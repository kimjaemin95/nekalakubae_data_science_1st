```python
population = pd.read_excel("../data/07_population_raw_data.xlsx", header=1)
population.fillna(method="pad", inplace=True)
population.head()
```

```python
population.info()
```

```python
# 컬럼 이름 변경
population.rename(columns={
    "행정구역(동읍면)별(1)":"광역시도",
    "행정구역(동읍면)별(2)":"시도",
    "계":"인구수",
}, inplace=True)
```

```python
population.head()
```

```python
population = population[population["시도"] != "소계"]
population.head()
```

```python
population.is_copy = False

population.rename(columns={
    "항목":"구분"
}, inplace=True)
```

```python
population.head()
```

```python
population["구분"].unique()
```

```python
population.loc[population["구분"] == "총인구수 (명)", "구분"] = "합계"
population.loc[population["구분"] == "남자인구수 (명)", "구분"] = "남자"
population.loc[population["구분"] == "여자인구수 (명)", "구분"] = "여자"
```

```python
population.head()
```

```python
# 소멸지역을 검사하기 위한 데이터 

population["20-39세"] = (
    population["20 - 24세"] +
    population["25 - 29세"] +
    population["30 - 34세"] +
    population["35 - 39세"]
)
population["65세이상"] = (
    population["65 - 69세"] +
    population["70 - 74세"] + 
    population["75 - 79세"] +
    population["80 - 84세"] +
    population["85 - 89세"] +
    population["90 - 94세"] +
    population["95 - 99세"] +
    population["100+"]
)
```

```python
population.tail()
```

```python
pop = population.pivot_table(index=["광역시도", "시도"], columns=["구분"], values=["인구수", "20-39세", "65세이상"])
pop
```

```python
# 소멸 비율 계산

pop["소멸비율"] = pop[("20-39세", "여자")] / (pop[("65세이상","합계")] / 2)
pop.tail()
```

```python
# 소멸위기지역 컬럼 생성

pop["소멸위기지역"] = pop["소멸비율"] < 1.0 
pop.head(10)
```

```python
# 소멸위기지역 조회

pop[pop["소멸위기지역"] == True].index.get_level_values(1)
```

```python
pop.reset_index(inplace=True)
```

```python
pop.head()
```

```python
pop.columns = ["".join(i) for i in pop.columns]
```

```python
pop
```

## 3. 지도 시각화를 위한 지역별 ID 만들기

```python
pop.info()
```

```python
pop.head()
```

```python
pop["시도"].unique()
```

```python
si_name = [None] * len(pop)
len(si_name)
```

```python
# 행정구 정리
tmp_gu_dict = {
    "수원": ["장안구", "권선구", "팔달구", "영통구"],
    "성남": ["수정구", "중원구", "분당구"],
    "안양": ["만안구", "동안구"],
    "안산": ["상록구", "단원구"],
    "고양": ["덕양구", "일산동구", "일산서구"],
    "용인": ["처인구", "기흥구", "수지구"],
    "청주": ["상당구", "서원구", "흥덕구", "청원구"],
    "천안": ["동남구", "서북구"],
    "전주": ["완산구", "덕진구"],
    "포항": ["남구", "북구"],
    "창원": ["의창구", "성산구", "진해구", "마산합포구", "마산회원구"],
    "부천": ["오정구", "원미구", "소사구"],
}
```

```python
pop.head()
```

```python
pop["광역시도"].unique()
```

```python
pop["시도"].unique()
```

#### (1) 일반 시 이름과 세종시, 광역시도 일반 구 정리

```python
pop.head(2)
```

```python
for idx, row in pop.iterrows():
    if row["광역시도"][-3:] not in ["광역시", "특별시", "자치시"]:
        si_name[idx] = row["시도"][:-1]
    elif row["광역시도"] == "세종특별자치시":
        si_name[idx] = "세종"
    else:
        if len(row["시도"]) == 2:
            si_name[idx] = row["광역시도"][:2] + " " + row["시도"]
        else:
            si_name[idx] = row["광역시도"][:2] + " " + row["시도"][:-1]
```

#### (2) 행정구

```python
for idx, row in pop.iterrows():
    if row["광역시도"][-3:] not in ["광역시", "특별시", "자치시"]:
        for key, values in tmp_gu_dict.items():
            if row["시도"] in values:
                if len(row["시도"]) == 2:
                    si_name[idx] = key + " " + row["시도"]
                
                elif row["시도"] in ["마산합포구", "마산회원구"]:
                    si_name[idx] = key + " " + row["시도"][2:-1]

                else:
                    si_name[idx] = key + " " + row["시도"][:-1]
```

### (3) 고성군

```python
for idx, row in pop.iterrows():
    if row["광역시도"][-3:] not in ["광역시", "특별시", "자치시"]:
        if row["시도"][:-1] == "고성" and row["광역시도"] == "강원도": 
            si_name[idx] = "고성(강원)"
        elif row["시도"][:-1] == "고성" and row["광역시도"] == "경상남도": 
            si_name[idx] = "고성(경남)"
```

```python
si_name
```

```python
pop["ID"] = si_name
```
