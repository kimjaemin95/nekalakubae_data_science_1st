- 해당 지역의 가임 여성 인구수 / (65세 이상의 노인 전체 인구수 / 2)

```python
def plot_text_simple(draw_korea):
    
    for idx, row in draw_korea.iterrows():
        id_nm = row["ID"]
        id_sp = id_nm.split()
        
        if len(id_sp) == 2:
            disp_name = "{}\n{}".format(id_sp[0], id_sp[1])
            
        elif id_nm[:2] == "고성":
            disp_name = "고성"
            
        else:
            disp_name = row["ID"]
            
        if len(disp_name.splitlines()[-1]) >= 3:
            fontsize, line_spacing = 9.5, 1.5
        else:
            fontsize, line_spacing = 11, 1.2
        
        # matplotlib에서 주석을 달기 위한 함수 모듈
        plt.annotate(
            text = disp_name,
            xy = (row["x"] + 0.5, row["y"] + 0.5),
            weight = "bold",
            fontsize = fontsize,
            linespacing = line_spacing,
            ha = "center", # 수평 정렬
            va = "center", # 수직 정렬
        )
    
```

```python
def simple_draw(draw_korea):
    plt.figure(figsize=(9, 13))
    
    plot_text_simple(draw_korea)
    
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs,ys, c="black", lw=1.5)
    
    plt.gca().invert_yaxis() # y축 기준으로 상하 대칭
    plt.axis("off")
    plt.tight_layout() # 주석 뭐식....
    plt.show()
```

```python
simple_draw(draw_korea)
```

#### 그림을 그리기 위한 데이터를 계산하는 함수
- 색상을 만들 때, 최소값을 흰색
- blocked_map : 인구현황(pop)
- target_data : 그리고 싶은 컬럼

```python
def get_data_info(target_data, blocked_map):
    
    data = blocked_map[target_data]
    
    white_label_min = (
        max(data) - min(data)
    ) * 0.25 + min(data)
    
    vmin = min(data)
    vmax = max(data)
    
    map_data = blocked_map.pivot_table(index="y", columns="x", values = target_data)

    return map_data, vmax, vmin, white_label_min
```

```python
def get_data_info_for_zero_center(target_data, blocked_map):
    
    data = blocked_map[target_data]
    
    white_label_min = 5
    
    tmp_max = max([np.abs(min(data)), np.abs(max(data))])
    vmin = -tmp_max
    vmax = tmp_max
    
    map_data = blocked_map.pivot_table(index="y", columns="x", values = target_data)

    return map_data, vmax, vmin, white_label_min
```

```python
def plot_text(target_data, blocked_map, white_label_min):
    
    for idx, row in blocked_map.iterrows():
        id_nm = row["ID"]
        id_sp = id_nm.split()
        
        if len(id_sp) == 2:
            disp_name = "{}\n{}".format(id_sp[0], id_sp[1])
            
        elif id_nm[:2] == "고성":
            disp_name = "고성"
            
        else:
            disp_name = row["ID"]
            
        if len(disp_name.splitlines()[-1]) >= 3:
            fontsize, line_spacing = 9.5, 1.5
        else:
            fontsize, line_spacing = 11, 1.2
        
        # 색상을 지정하기 위함
        anno_color = "white" if np.abs(row[target_data]) > white_label_min else "black"
        
        
        # matplotlib에서 주석을 달기 위한 함수 모듈
        plt.annotate(
            text = disp_name,
            xy = (row["x"] + 0.5, row["y"] + 0.5),
            weight = "bold",
            color=anno_color,
            fontsize = fontsize,
            linespacing = line_spacing,
            ha = "center", # 수평 정렬
            va = "center", # 수직 정렬
        )
    
```

```python
def draw_korea_func(target_data, blocked_map, cmap_name, zero_center=False): 
    
    if zero_center:
        masked_map_data, vmax, vmin, white_label_min = get_data_info_for_zero_center(target_data, blocked_map)
    else: 
        masked_map_data, vmax, vmin, white_label_min = get_data_info(target_data, blocked_map)
    
    plt.figure(figsize=(9, 13))
    plt.pcolor(masked_map_data, vmin=vmin, vmax=vmax, cmap=cmap_name, edgecolor="#aaaaaa", linewidth=0.5)
    
    plot_text(target_data, blocked_map, white_label_min)
    
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs,ys, c="black", lw=1.5)
    
    plt.gca().invert_yaxis() # y축 기준으로 상하 대칭
    plt.axis("off")
    plt.tight_layout() # 주석 뭐식....
    cb = plt.colorbar(shrink=0.1, aspect=10)
    cb.set_label(target_data)
    plt.show()
```
