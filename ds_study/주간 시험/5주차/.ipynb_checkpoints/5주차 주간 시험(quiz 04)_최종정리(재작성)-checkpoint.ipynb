### 히스토리 데이터 추출 함수 선언

```python
def set_df(df):
    '''
    입력 받은 데이터프레임을 복사하고
    히스토리 데이터 반환
    '''
    df_cp = df.copy()
    df_cp.index = df_cp["Country/Region"]

    del df_cp["Province/State"]
    del df_cp["Country/Region"]
    del df_cp["Lat"]
    del df_cp["Long"]
    del df_cp["Section"]

    return df_cp
```

```python
confirmed_df_5_cp = set_df(confirmed_df_5)
deaths_df_5_cp = set_df(deaths_df_5)

confirmed_df_5_cp.head()
```

### 데이터 시각화 함수 선언

```python
def draw_grahp(df, title):
    f = plt.figure(figsize=(15,5))
    
    for num in range(1, 3):
        
        # 그래프 위치 및 데이터 설정
        p = f.add_subplot(1,2,num)
        for idx, rows in df.iterrows():
            if (num > 1) and (idx == "US"):
                continue
            p.plot([idx for idx, _ in enumerate(rows.index)], list(rows), label=idx)
            
        # 제목 설정 
        if num > 1: title += "(미국 제외)"
            
        # X축 라벨
        xlabel = "2020년 1월 22일 ~ 최신"
        
        p.set_title(title)
        p.set_xlabel(xlabel)
        p.legend()
        
    plt.show()
```

### 국가별 확진자, 사망자 히스토리 데이터로 방역 대응 평가가 가능할까?
- 전체 인구수가 많은 국가에서 확진자수와 사망자수가 더 많은 것은 비율상 자연스럽다.
- 국가 문화, 유동 인구, 지형, 인구 밀집, 보건 의료 정책 등의 다양한 지표가 필요하다.

### 코로나 19 방역 대응 평가 분석 고려사항
- 인구수 대비 비율
    - 국가별 (환진, 사망 인구수) / (전체 인구수) 
    - 국가별 10만명당 확진 또는 사망 인구수 확인

### 코로나 19 방역 대응 평가 대상 국가의 전체 인구수 확인

```python
# 이탈리아 : 6,036만 7,471명
# 독일 : 8,390만 471명
# 일본 : 1억 2,605만 796명
# 대한민국 : 5,182만 1,669명
# 미국 : 3억 3,291만 5,074명
italy_p = 60367471 ; germany_p = 83900471 ; japan_p = 126050796 ; korea_p = 51821669 ; us_p = 332915074
```

### 코로나 19 확진자, 사망자 누적 데이터와 국가별 전체 인구수 데이터프레임 생성

```python
# 누적확진자, 누적사망자 데이터 추출(최근 데이터)
data = {
    "누적확진자" : confirmed_df_5_cp[confirmed_df_5_cp.columns[-1]], 
    "누적사망자" : deaths_df_5_cp[deaths_df_5_cp.columns[-1]]
}

# 최근(누적) 데이터 프레임 생성
covid_curr_df = pd.DataFrame(data)
covid_curr_df

# 분석 대상 국가의 인구수 데이터 추가
covid_curr_df.loc["Germany", "인구수"] = germany_p
covid_curr_df.loc["Italy", "인구수"] = italy_p
covid_curr_df.loc["Japan", "인구수"] = japan_p
covid_curr_df.loc["Korea, South", "인구수"] = korea_p
covid_curr_df.loc["US", "인구수"] = us_p

# 인구수 데이터 정수형 변환
covid_curr_df["인구수"] = covid_curr_df["인구수"].astype("int64")

covid_curr_df
```

### 국가별 확진률, 사망률 데이터 생성

```python
covid_curr_df["확진률"] = round(covid_curr_df["누적확진자"] / covid_curr_df["인구수"] * 100, 5)
covid_curr_df["사망률"] = round(covid_curr_df["누적사망자"] / covid_curr_df["누적확진자"] * 100, 5)

covid_curr_df
```

### 국가별 인구수대비 확진률, 사망률 데이터 시각화

```python
covid_curr_df[["확진률", "사망률"]].plot.bar(figsize=(12,5), title="국가별 인구수대비 확진률/사망률");
```

### 시계열 분석을 통한 국가별 코로나 19 트렌드 확인 및 주기별 특성 관찰

```python
def covid_prophet(country:str, df):
    '''
    나라별 코로나 19 데이터의 트렌드와 주기별 특성을 관찬
    country : 분석 나라 입력
    df : 확진자 또는 사망자 데이터프레임 입력
    '''
    # 분석 데이터 전처리
    data = {
        "ds" : df.columns,
        "y" : df[df.index == country].values.tolist()[0]
    }
    df = pd.DataFrame(data)
    
    # Prophet 모델 학습
    m = Prophet(yearly_seasonality=True, daily_seasonality=True)
    m.fit(df)
    future = m.make_future_dataframe(periods=10)
    
    # Prophet 데이터 예측
    forecast = m.predict(future)
    # forecast_source = forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]]
    
    # Prophet 시계열 데이터 시각화
    fig = m.plot(forecast);
    a = add_changepoints_to_plot(fig.gca(), m, forecast)
    
    # Prophet Trend 파악
    m.plot_components(forecast);
```

#### covid_prophet(국가, 확진자 또는 사망자 데이터프레임)

### 대한민국 코로나 19 확진자 트렌드 파악

```python
covid_prophet("Korea, South", confirmed_df_5_cp)
```

### 대한민국 코로나 19 사망자 트렌드 파악

```python
covid_prophet("Korea, South", deaths_df_5_cp)
```

### 인구 10만명당 감염자, 사망자 확인
- 같은 인구수일 때 평균 코로나 19 감염자와 사망자의 데이터를 확인

```python
standard = 100000
covid_curr_df["인구10만명당_감염_발생자"] = round((covid_curr_df["누적확진자"] * standard)/covid_curr_df["인구수"], 0)
covid_curr_df["인구10만명당_사망_발생자"] = round((covid_curr_df["누적사망자"] * standard)/covid_curr_df["인구수"], 0)

covid_curr_df
```

### 인구 10만명당 감염, 사망 발생자 데이터 시각화용 함수 선언

```python
def draw_buble(df):

    # scatter()를 이용한 원 생성
    ax = df.plot.scatter(
        title="인구 10만명당 코로나19 확진자, 사망자 시각화",
        x="인구10만명당_감염_발생자",
        y="인구10만명당_사망_발생자",
        s=df["인구10만명당_감염_발생자"],
        c=df["인구10만명당_사망_발생자"],
        colormap='hot', # colormap = "viridis",
        alpha = 0.8,
        sharex = False,
        figsize = (10,5)
    )

    # 국가 이름 표기
    for idx, country in enumerate(df.index):
        ax.annotate(
            country, 
            (df.인구10만명당_감염_발생자.iat[idx], df.인구10만명당_사망_발생자.iat[idx]),
            ha = "center", # 수평 정렬
            va = "center", # 수직 정렬
        )

    # 시각화
    plt.show()
```

### 인구 10만명당 감염, 사망 발생자 데이터 시각화
- 버블차트 이용

```python
draw_buble(covid_curr_df)
```

### 세계 보건산업 시장규모 데이터 선언

```python
# 데이터 선언
df_medical = pd.read_csv("./세계_보건산업_시장규모(문제7번_자료).csv", encoding="utf-8", sep=",", index_col=1)

# 분석 대상 국가 선별
df_medical = df_medical.loc[["독일", "이탈리아", "일본", "대한민국", "미국"]]
df_medical.index = ["Germany", "Italy", "Japan", "Korea, South", "US"]
df_medical.rename(columns={"순위":"보건시장규모_순위"}, inplace=True)
df_medical
```

### 인구 10만명당 확진, 사망 데이터와 보건시장 규모 데이터의 정규화
