### 코로나 19 감염 대응 평가 분석 대상 국가 선정
- 한국, 미국, 이탈리아, 독일, 일본
- G20에 포함 되어 있는 국가이면서, 데이터에 도시(주) 별로 구분 되어 있지 않은 국가들

```python
# 확진자
confirmed_df_5 = confirmed_df.copy()[ 
    (confirmed_df["Country/Region"] == "Korea, South") | 
    (confirmed_df["Country/Region"] == "US") |
    (confirmed_df["Country/Region"] == "Italy") |
    (confirmed_df["Country/Region"] == "Germany") |
    (confirmed_df["Country/Region"] == "Japan")
]
confirmed_df_5["Section"] = "Confirmed"
```

```python
confirmed_df_5
```

```python
# 사망자
deaths_df_5 = deaths_df.copy()[ 
    (confirmed_df["Country/Region"] == "Korea, South") | 
    (confirmed_df["Country/Region"] == "US") |
    (confirmed_df["Country/Region"] == "Italy") |
    (confirmed_df["Country/Region"] == "Germany") |
    (confirmed_df["Country/Region"] == "Japan")
]
deaths_df_5["Section"] = "Deaths"
```

```python
deaths_df_5
```

### 히스토리 데이터 추출 함수 선언

```python
def set_df(df):
    '''
    입력 받은 데이터프레임을 복사하고
    히스토리 데이터 반환
    '''
    df_cp = df.copy()
    df_cp.index = df_cp["Country/Region"]

    del df_cp["Province/State"]
    del df_cp["Country/Region"]
    del df_cp["Lat"]
    del df_cp["Long"]
    del df_cp["Section"]

    return df_cp
```

```python
confirmed_df_5_cp = set_df(confirmed_df_5)
deaths_df_5_cp = set_df(deaths_df_5)

confirmed_df_5_cp.head()
```

### 국가별 코로나 19 확진자, 사망자 히스토리 데이터 시각화

```python
def draw_grahp(df, title):
    f = plt.figure(figsize=(15,5))
    
    for num in range(1, 3):
        
        # 그래프 위치 및 데이터 설정
        p = f.add_subplot(1,2,num)
        for idx, rows in df.iterrows():
            if (num > 1) and (idx == "US"):
                continue
            p.plot([idx for idx, _ in enumerate(rows.index)], list(rows), label=idx)
            
        # 제목 설정 
        if num > 1: title += "(미국 제외)"
            
        # X축 라벨
        xlabel = "2020년 1월 22일 ~ 최신"
        
        p.set_title(title)
        p.set_xlabel(xlabel)
        p.legend()
        
    plt.show()
```

```python
draw_grahp(confirmed_df_5_cp, "국가별 누적 확진자 수")
```

```python
draw_grahp(deaths_df_5_cp, "국가별 누적 사망자 수")
```

### 코로나 19 방역 대응 평가 대상 국가의 전체 인구수 확인

```python
# 이탈리아 : 6,036만 7,471명
# 독일 : 8,390만 471명
# 일본 : 1억 2,605만 796명
# 대한민국 : 5,182만 1,669명
# 미국 : 3억 3,291만 5,074명
italy_p = 60367471 ; germany_p = 83900471 ; japan_p = 126050796 ; korea_p = 51821669 ; us_p = 332915074
```

### 코로나 19 확진자 누적 현황 및 전체 인구수 데이터 전처리

```python
# 가장 마지막 날짜의 데이터
confirmed_5_curr = pd.DataFrame(confirmed_df_5_cp[confirmed_df_5_cp.columns[-1]])

# 국가별 전체 인구수 데이터
confirmed_5_curr["인구수"] = np.nan
confirmed_5_curr.loc["Germany", "인구수"] = germany_p
confirmed_5_curr.loc["Italy", "인구수"] = italy_p
confirmed_5_curr.loc["Japan", "인구수"] = japan_p
confirmed_5_curr.loc["Korea, South", "인구수"] = korea_p
confirmed_5_curr.loc["US", "인구수"] = us_p
```
