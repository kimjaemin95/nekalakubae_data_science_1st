```python
import numpy as np

x = np.arange(0, 12, 0.01)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```
